```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
import pandas as pd
import glob, os
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
path = 'scores'
```

```python
# combine all files into one df
all_files = glob.glob(os.path.join(path, "*.csv")) 
df_from_each_file = (pd.read_csv(f) for f in all_files)
df   = pd.concat(df_from_each_file, ignore_index=True)
df.drop(['Unnamed: 0'],axis=1, inplace=True)
```

```python
df
```

```python
df['feture_extraction'] = df[['CountVectorizer','TfidfVectorizer']].idxmax(axis=1)
```

```python
df.drop(['CountVectorizer','TfidfVectorizer'],axis=1,inplace=True)
```

```python
df
```

## separate by model

```python
df.model.value_counts()
```

```python
mnb = df.model == 'MultinomialNB'
logr = df.model == 'LogisticRegression'
lsvc = df.model == 'LinearSVC'
```

```python
mnbayes = df[mnb].dropna(axis=1)
mnbayes
```

```python
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
```

```python
sns.set(rc={'figure.figsize':(10,5)})
```

```python
ax = sns.lineplot(x="text", y="score", hue="feture_extraction",data=mnbayes)
plt.title('MultinomialNB Score')
plt.savefig("MNBayes.png")
plt.show()
```

```python
ax = sns.lineplot(x="alpha", y="score", hue="text",data=mnbayes)
plt.title('MultinomialNB Score')
plt.savefig("MNBayes_a.png")

plt.show()
```

```python
logrdf = df[logr].dropna(axis=1)
logrdf
```

```python
ax = sns.lineplot(x="text", y="score", hue="feture_extraction",data=logrdf)
plt.title('LogisticRegression')
plt.savefig("logr.png")
plt.show()
```

```python
ax = sns.lineplot(x="C_value", y="score", hue="text",data=logrdf)
plt.title('LogisticRegression')
plt.savefig("logr_cval.png")

plt.show()
```

```python
ax = sns.lineplot(x="max_df", y="score", hue="text",data=logrdf)
plt.title('LogisticRegression')
plt.savefig("logr_df.png")

plt.show()
```

```python
ax = sns.lineplot(x="ngram_range", y="score", hue="text",data=logrdf)
plt.title('LogisticRegression')
plt.savefig("logr_ngram.png")

plt.show()
```
