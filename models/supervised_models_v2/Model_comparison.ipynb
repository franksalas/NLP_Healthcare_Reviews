```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
path = 'yelp_data/health_text_model_results.csv'

df = pd.read_csv(path)
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df.head()
```

## Change fold val for better plotting

```python
df.fold_idx.replace(4,5,inplace=True)
```

```python
df.fold_idx.replace(3,4,inplace=True)
```

```python
df.fold_idx.replace(2,3,inplace=True)
```

```python
df.fold_idx.replace(1,2,inplace=True)
```

```python
df.fold_idx.replace(0,1,inplace=True)
```

```python
df.plot()
```

```python
sns.lineplot(x="fold_idx", y="accuracy", hue="model_name",data=df)
plt.yticks(fontsize=15)

plt.ylabel('')
# set the title of the graph
plt.title('Model Comparison',fontsize=15)
plt.xlabel('Ratings', fontsize=15)
plt.show()
```

```python
sns.set_context("talk")
sns.set(rc={'figure.figsize':(10,5)})

sns.lineplot(x="fold_idx", y="accuracy", hue="model_name",data=df,legend=False)
```
