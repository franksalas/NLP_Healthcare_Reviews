````python
title = 'Top 10 Cities'
x_text = 'City'
y_text = 'Total Reviews'
#Get the distribution
x=df_full.city.value_counts()[:10]
#x=x.sort_index()

#plot
plt.figure(figsize=(10, 6))
ax= sns.barplot(x.index, x.values, alpha=0.8)
plt.title(title)
plt.ylabel(y_text, fontsize=12)
plt.xlabel(x_text, fontsize=12)

#adding the text labels
rects = ax.patches
labels = x.values
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height + 5, label, ha='center', va='bottom')

# save img
#plt.savefig('{}/{}.png'.format(img_directory,title.replace(" ", "_")))
plt.savefig('{}/{}.svg'.format(img_directory,title.replace(" ", "_"),dpi=1200))
````

## Sentiment Analysis

````python
# Let's plot the results
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [10, 6]
title = 'Sentiment Analysis: Star Review'
for i in df_text.stars.unique():
    x = df_text.groupby(['stars']).get_group(i)['polarity'].median()
    y = df_text.groupby(['stars']).get_group(i)['subjectivity'].median()
    plt.scatter(x, y, color='blue')
    plt.text(x+.001, y+.001,i, fontsize=10)
    #plt.xlim(-1,1)
    #plt.ylim(0,1)
plt.title(title, fontsize=20)
#plt.plot(np.arange(0,6), np.zeros(6)+.5,'r--')
plt.xlabel('<-- Negative\t\t\tPositive -->'.expandtabs(), fontsize=15)
plt.ylabel('<-- Facts\t\t\tOpinions -->'.expandtabs(), fontsize=15)
# save img
plt.savefig('{}/{}.svg'.format(img_directory,title.replace(" ", "_"),dpi=1200))
plt.show()
````

````python
# Let's plot the results
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [10, 6]
title = 'Sentiment Analysis:Health Business '
for i in df_text.health_business.unique():
    x = df_text.groupby(['health_business']).get_group(i)['polarity'].median()
    y = df_text.groupby(['health_business']).get_group(i)['subjectivity'].median()
    plt.scatter(x, y, color='blue')
    plt.text(x+.001, y+.001,i, fontsize=10)
    #plt.xlim(-1,1)
    #plt.ylim(0,1)
plt.title(title, fontsize=20)
plt.xlabel('<-- Negative\t\t\tPositive -->'.expandtabs(), fontsize=15)
plt.ylabel('<-- Facts\t\t\tOpinions -->'.expandtabs(), fontsize=15)
# save img
#plt.savefig('{}/{}.png'.format(img_directory,title.replace(" ", "_")))
plt.savefig('{}/{}.svg'.format(img_directory,title.replace(" ", "_"),dpi=1200))

plt.show()
````

````python
title = 'Sentiment Analysis: Subjectivity'
df_text.groupby(['stars','health_business']).median()['subjectivity'].unstack().plot(colormap='Paired')
plt.title(title, fontsize=20)
plt.xlim([1,5])
plt.plot(np.arange(1,6), np.zeros(5)+.5,'r--')

plt.xticks([1,2,3,4,5])
ylab = '<-- Facts\t\t\tOpinions -->'.expandtabs()
plt.xlabel('Stars', fontsize=15)
plt.ylabel(ylab, fontsize=15)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.tight_layout()

# save img
#plt.savefig('{}/{}.png'.format(img_directory,title.replace(" ", "_")))
plt.savefig('{}/{}.svg'.format(img_directory,title.replace(" ", "_"),dpi=1200))

plt.show()
````

````python
title = 'Sentiment Analysis: Polarity'
df_text.groupby(['stars','health_business']).median()['polarity'].unstack().plot(colormap='Paired')
plt.title(title, fontsize=20)
plt.xlim([1,5])
plt.xticks([1,2,3,4,5])
#plt.plot(np.arange(1,6), np.zeros(5),'r--')
plt.axhline(0, color='k')

#plt.ylim([-1,1])
plt.xlabel('Stars', fontsize=15)
plt.ylabel('Polarity', fontsize=15)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.tight_layout()

# save img
#plt.savefig('{}/{}.png'.format(img_directory,title.replace(" ", "_")))
plt.savefig('{}/{}.svg'.format(img_directory,title.replace(" ", "_"),dpi=1200))
plt.show()
````
